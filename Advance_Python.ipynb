```python
import random
import string

with open("q1_1000_lines.txt", "w") as f:
    for _ in range(1000):
        line = ''.join(random.choices(string.ascii_letters + string.digits, k=80))
        f.write(line + '\n')
```

```python
def create_5mb_file(filename):
    with open(filename, 'w') as f:
        while f.tell() < 5 * 1024 * 1024:  # 5MB
            line = ''.join(random.choices(string.ascii_letters + string.digits, k=100))
            f.write(line + '\n')

create_5mb_file("q2_5mb_file.txt")
```

```python
for i in range(10):
    create_5mb_file(f"q3_file_{i+1}.txt")
```

```python
def create_gb_file(filename, size_gb):
    with open(filename, 'w') as f:
        line = ''.join(random.choices(string.ascii_letters + string.digits, k=100)) + '\n'
        total_lines = int((size_gb * 1024 * 1024 * 1024) / len(line))
        for _ in range(total_lines):
            f.write(line)

# Uncomment the following to run (slow & needs disk space)
# for size in range(1, 6):
#     create_gb_file(f"q4_file_{size}GB.txt", size)
```

```python
def convert_uppercase(input_file, output_file):
    with open(input_file, 'r') as infile:
        content = infile.read().upper()
    with open(output_file, 'w') as outfile:
        outfile.write(content)

# Example usage
# convert_uppercase("q4_file_1GB.txt", "q5_upper_1GB.txt")
```

```python
import threading

def threaded_uppercase(file_index):
    input_file = f"q4_file_{file_index}GB.txt"
    output_file = f"q6_upper_{file_index}GB.txt"
    convert_uppercase(input_file, output_file)

threads = []
for i in range(1, 6):
    t = threading.Thread(target=threaded_uppercase, args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
def download_and_convert(video_url, index):
    yt = YouTube(video_url)
    stream = yt.streams.filter(progressive=True, file_extension='mp4').first()
    stream.download(filename=f"video_{index}.mp4")
    clip = VideoFileClip(f"video_{index}.mp4")
    clip.audio.write_audiofile(f"audio_{index}.mp3")

urls = [YouTube(video.watch_url).watch_url for video in Search("Machine Learning").results[:10]]
threads = []

for i, url in enumerate(urls):
    t = threading.Thread(target=download_and_convert, args=(url, i))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import os
import requests

# Download the google_images_download.py file
url = "https://raw.githubusercontent.com/Joeclinton1/google-images-download/master/google_images_download/google_images_download.py"
response = requests.get(url)
with open("google_images_download.py", "w") as f:
    f.write(response.text)

# Patch the file to fix the chrome_options issue
with open("google_images_download.py", "r") as f:
    lines = f.readlines()

patched_lines = []
for line in lines:
    if "browser = webdriver.Chrome(chromedriver, chrome_options=options)" in line:
        patched_lines.append(line.replace("chrome_options=options", "options=options"))
    else:
        patched_lines.append(line)

with open("google_images_download.py", "w") as f:
    f.writelines(patched_lines)

# Now we can import and use the patched version
from google_images_download import google_images_download

print("Patched google_images_download.py and imported successfully.")
```

```python
from PIL import Image
import os

response = google_images_download.googleimagesdownload()
response.download({"keywords": "dog", "limit": 10, "output_directory": "dogs"}) # Reduced limit for faster execution

def resize_images():
    path = "downloads/dog" # Corrected path
    if not os.path.exists(path):
        print(f"Directory not found: {path}")
        return
    for img in os.listdir(path):
        try:
            img_path = os.path.join(path, img)
            # Check if it's a file before trying to open
            if os.path.isfile(img_path):
                im = Image.open(img_path)
                im_resized = im.resize((int(im.width * 0.5), int(im.height * 0.5)))
                # Create a directory for resized images if it doesn't exist
                resized_path = os.path.join(path, "resized")
                os.makedirs(resized_path, exist_ok=True)
                im_resized.save(f"{resized_path}/resized_{img}")
        except Exception as e:
            print(f"Error processing image {img}: {e}")
            continue

resize_images()
```
